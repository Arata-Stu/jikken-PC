```python
import matplotlib.pyplot as plt
from PIL import Image

# 画像ファイルのパス
image_path = 'image/sponge/sponge_lack3.png' # ここに画像ファイルのパスを指定してください。

# 画像の読み込み
img = Image.open(image_path)

# 画像を表示
plt.imshow(img)

# グラフに軸を追加
plt.xlabel('X axis')
plt.ylabel('Y axis')

# タイトルを追加
plt.title('Image Display with Axes')

# 画像表示
plt.show()
```
